```python
#A2. Detect the points of zero crossing in the first derivative signal. Compare the average length
#between two consecutive zero crossings for speech and silence regions. Observe the pattern.
z=librosa.zero_crossings(first_derivative)
print(z)
zc=np.nonzero(z)
zc
```

```python
print(len(first_derivative))
```

```python
import numpy as np
import soundfile as sf 
zero_crossing_intervals = np.diff(z)
print(zero_crossing_intervals)
threshold_value = 0.05 
speech_regions = zero_crossing_intervals[zero_crossing_intervals > threshold_value * sr1]
silence_regions = zero_crossing_intervals[zero_crossing_intervals <= threshold_value * sr1]
average_speech_length = np.mean(speech_regions) / sr1
average_silence_length = np.mean(silence_regions) / sr1

print("Average length between the two consecutive zero crossings in speech regions is:", average_speech_length, "seconds")
print("Average length between the two consecutive zero crossings in silence regions is:", average_silence_length, "seconds")
```

```python
# To plot the Speech and Silent Region
import librosa
import librosa.display
import matplotlib.pyplot as plt
# Calculate short-term energy
energy = librosa.feature.rms(y=y)
# Set an energy threshold
threshold = 0.01
# Identify speech and silence regions
speech_regions = energy > threshold
silence_regions = energy <= threshold
# Plot the results
plt.figure(figsize=(12, 8))
librosa.display.waveshow(y, alpha=0.5,color='g')
librosa.display.specshow(energy, x_axis='time', sr=sr1, color='r', alpha=0.75)
plt.title('Speech and Silence Regions')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
#A3. Speak 5 of your favorite words. Observe the length of the speech signals. Compare the lengths of
#your spoken words with those of your project team-mate
y1, sr2= librosa.load("colors.wav")
y2, sr3= librosa.load("lab2.wav")
plt.figure(figsize=(14, 5))
plt.subplot(1,2,1)
librosa.display.waveshow(y1, color= "pink")
plt.title('My Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(1, 2, 2)
librosa.display.waveshow(y2,color='blue')
plt.title("Teammates' Signal")
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

#signal length of the my signal and my teammate's
duration1 = librosa.get_duration(y=y1, sr=sr2)
duration2 = librosa.get_duration(y=y2, sr=sr3)
print(duration1)
print(duration2)

# calculating the difference in the sample rate of my signal with the teammate's signal 
print("sampling rate difference: ",sr2-sr3) 
```

```python
# visulising the signal difference of my signal with respect to the teammates signal
librosa.display.waveshow(y1, sr=sr2,color='pink')
librosa.display.waveshow(y2, sr=sr3, color= "blue")
plt.title('Signal difference')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```
