```python
D = librosa.stft(y)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)  #converting the magnitude spectrogram to the decibel (dB) scale for visualisation
S_db.shape
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
img = librosa.display.specshow(S_db,
                              x_axis='time',
                              y_axis='log',
                              ax=ax)
ax.set_title('Spectogram Example', fontsize=10)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```
