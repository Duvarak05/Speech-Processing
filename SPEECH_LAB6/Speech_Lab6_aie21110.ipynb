```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pylab as plt
from IPython.display import Audio
```

```python
#A1. Take a portion of your recorded signal which represents a vowel sound. Perform FFT on the
#signal snippet and observe the amplitude spectrum. Repeat the same for a few vowel sounds.
y, sr = librosa.load("a.wav")
librosa.display.waveshow(y,color='pink') 
```

```python
Audio(data=y, rate=sr)
```

```python
yt, index= librosa.effects.trim(y,top_db=10) 
```

```python
Audio(data=yt, rate=sr)
```

```python
librosa.display.waveshow(yt,color="pink") #wave of letter a
```

```python
arr = np.fft.fft(yt)
plt.plot(np.abs(arr))
plt.title("Absolute part of the spectral components of letter a")
```

```python
#A3. Repeat A2 for few slices of silence & non-voiced portions of the recorded speech signal
y_silence, srs = librosa.load("silence.wav")
#removing the initial noise
yt, index= librosa.effects.trim(y_silence,top_db=10) 
librosa.display.waveshow(yt,color='blue') #wave of letter e
arr_silence = np.fft.fft(yt)
```

```python
import matplotlib.pyplot as plt
plt.plot(np.abs(arr_silence[0:len(arr_silence)//2]))
plt.title("Absolute part of the silence or non voiced sound")
```

```python
#A4. Now you have acquainted yourself with spectral amplitudes of various consonants and vowel-basedphonemes. 
#Generate the spectrogram of thesignal and observe the change points of the signals with associatedspeech segments. 
#Observe to identify the consonants and vowels from the spectrogram.
hop_size = int(0.02 *sr3 / 2)
window_length=int(sr3*0.02) # windows of size 20ms
stft_matrix = librosa.stft(y3, n_fft=window_length, hop_length=window_length)
# Convert complex spectrogram to magnitude spectrogram
magnitude_stft = np.abs(stft_matrix)
# Plot the magnitude spectrogram as a heatmap
plt.figure(figsize=(10, 6))
librosa.display.specshow(librosa.amplitude_to_db(magnitude_stft, ref=np.max), sr=sr3, hop_length=window_length, x_axis='time', y_axis='hz')
plt.colorbar(format='%+2.0f dB')
plt.title('Magnitude Spectrogram of vowel speech Signal')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.tight_layout()
plt.show()
```
